### 전국 초등학교 데이터를 구합니다

```python
import time

# 실행 전 시간 기록
start_time = time.time()

columns = [
    'ATPT_OFCDC_SC_CODE',
    'ATPT_OFCDC_SC_NM',
    'SD_SCHUL_CODE',
    'SCHUL_NM',
    'ENG_SCHUL_NM',
    'SCHUL_KND_SC_NM',
    'LCTN_SC_NM',
    'JU_ORG_NM',
    'FOND_SC_NM',
    'ORG_RDNZC',
    'ORG_RDNMA',
    'ORG_RDNDA',
    'ORG_TELNO',
    'HMPG_ADRES',
    'COEDU_SC_NM',
    'ORG_FAXNO',
    'HS_SC_NM',
    'INDST_SPECL_CCCCL_EXST_YN',
    'HS_GNRL_BUSNS_SC_NM',
    'SPCLY_PURPS_HS_ORD_NM',
    'ENE_BFE_SEHF_SC_NM',
    'DGHT_SC_NM',
    'FOND_YMD',
    'FOAS_MEMRD',
    'LOAD_DTM'
]

new_columns = [
    '시도교육청코드',  
    '시도교육청명',
    '행정표준코드',
    '학교명',
    '영문학교명',
    '학교종류명',
    '시도명',
    '관할조직명',
    '설립명',
    '도로명우편번호',
    '도로명주소',
    '도로명상세주소',
    '전화번호',
    '홈페이지주소',
    '남녀공학구분명',
    '팩스번호',
    '고등학교구분명',
    '산업체특별학급존재여부',
    '고등학교일반전문구분명',
    '특수목적고등학교계열명',
    '입시전후기구분명',
    '주야구분명',
    '설립일자',
    '개교기념일',
    '수정일자'
]

# API 키 세팅
api_key = 'eb7332a959474e749c25c0a784cfbedd'

url = 'https://open.neis.go.kr/hub/schoolInfo'

# 데이터를 담는 리스트
df_total = []

for i in range(1,8):
    
    #요청변수 파라미터 설정
    params = '?' + \
             'KEY=' + api_key + '&' + \
             'Type=' + 'xml' + '&' + \
             'pIndex=' + str(i) + '&' + \
             'pSize=' + '1000' + '&' + \
             'SCHUL_NM=' + "초등학교"
    
    res = requests.get(url + params)
    soup = bs(res.text,'xml')
    items = soup.find_all('row')
    
    # 수집된 데이터를 데이터 프레임 형식으로 변환
    for j in range(len(items)):
        df_raw = []
        df_list = items[j].find_all()
        for idx ,column in enumerate(columns):
            # 컬럼 데이터가 없는 경우 처리
            try:
                df_raw.append(df_list[idx].text)
            except:
                df_raw.append('없음')
        # 데이터 행 하나를 더해줍니다        
        df_total.append(df_raw)

    print('-' * 100)
    
total = pd.DataFrame(df_total)
total.columns = new_columns    
    
# 실행 후 시간 기록
end_time = time.time()    
    
# 실행 시간 계산
execution_time = end_time - start_time
print(f"실행 시간: {execution_time} 초")   
    
total.reset_index(drop=True,inplace=True)
total
```

```python
directory = '0_data/school'

if not os.path.exists(directory):
    os.makedirs(directory)

file_name = f"{directory}/elementary_school.csv"

total.to_csv(file_name, index = False)
```
